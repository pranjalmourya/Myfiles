```python
from threading import local
from pyspark.sql import SparkSession
spark = (
     SparkSession
    .builder
    .appName("Sort_union_agg")
    .master("local[*]")
    .getOrCreate()
      )
```

```python
Emp_schema = "employee_id string, department_id string, name string, age string, gender string, salary string, hire_date string"
```

```python
emp1 = spark.createDataFrame(emp_data_1,schema = Emp_schema)
emp2 = spark.createDataFrame(emp_data_2,schema = Emp_schema)
```

```python
emp1.show()
emp2.show()
```

```python
Emp_union = emp1.union(emp2)
Emp_union.show()
```

```python
from pyspark.sql.functions import desc, asc,col
Emp_sorter = Emp_union.orderBy(col("salary").desc())
Emp_sorter.show()
```
